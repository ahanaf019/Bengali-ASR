```python
import os
from pydub import AudioSegment

def get_audio_length(folder_path):
    audio_lengths = {}
    for filename in os.listdir(folder_path):
        if filename.endswith('.mp3') or filename.endswith('.wav'):
            file_path = os.path.join(folder_path, filename)
            audio = AudioSegment.from_file(file_path)
            length_seconds = len(audio) / 1000  # Convert milliseconds to seconds
            audio_lengths[filename] = length_seconds
    return audio_lengths

folder_path = '/media/ahanaf/media-1/Datasets/ben10/ben10/16_kHz_train_audio'
audio_lengths = get_audio_length(folder_path)

for filename, length_seconds in audio_lengths.items():
    print(f"{filename}: {length_seconds} seconds")
```

```python
import wave
from tqdm import tqdm
```

```python
def get_bit_depth(file_path):
    with wave.open(file_path, 'rb') as audio_file:
        bit_depth = audio_file.getsampwidth() * 8
    return bit_depth

def count_bit_depths(folder_path):
    bit_depth_count = {'16-bit': 0, '8-bit': 0, '32-bit': 0}
    for filename in tqdm(os.listdir(folder_path)):
        try:
            if filename.endswith('.wav'):
                file_path = os.path.join(folder_path, filename)
                bit_depth = get_bit_depth(file_path)
                if bit_depth == 16:
                    bit_depth_count['16-bit'] += 1
                elif bit_depth == 8:
                    bit_depth_count['8-bit'] += 1
                elif bit_depth == 32:
                    bit_depth_count['32-bit'] += 1
        except Exception as e:
            print(filename, e)
    return bit_depth_count

folder_path = '/media/ahanaf/media-1/Datasets/ben10/ben10/16_kHz_train_audio'
bit_depth_count = count_bit_depths(folder_path)

print("Number of 16-bit audio files:", bit_depth_count['16-bit'])
print("Number of 8-bit audio files:", bit_depth_count['8-bit'])
print("Number of 32-bit audio files:", bit_depth_count['32-bit'])
```

```python
import os
import wave
from pydub import AudioSegment

most_wanted = []

def get_bit_depth(file_path):
    with wave.open(file_path, 'rb') as audio_file:
        bit_depth = audio_file.getsampwidth() * 8
    return bit_depth

def convert_to_16_bit(audio_file):
    return audio_file.set_sample_width(2)

def process_audio_files(folder_path):
    for filename in os.listdir(folder_path):
        try:
            if filename.endswith('.wav'):
                file_path = os.path.join(folder_path, filename)
                bit_depth = get_bit_depth(file_path)
                if bit_depth == 32:
                    # Convert 32-bit audio to 16-bit
                    audio = AudioSegment.from_file(file_path, format='wav')
                    audio = convert_to_16_bit(audio)
                    # Save the converted audio
                    output_path = os.path.splitext(file_path)[0] + '.wav'
                    audio.export(output_path, format='wav')
                    print(f"Converted {filename} to 16-bit and saved as {output_path}")
        except Exception as e:
            print(e, filename)
            most_wanted.append(filename)
            pass

folder_path = '/media/ahanaf/media-1/Datasets/ben10/ben10/16_kHz_train_audio'
process_audio_files(folder_path)
```

```python
import os
from pydub import AudioSegment
from pydub.utils import mediainfo

def convert_mulaw_to_pcm(file_path):
    # Load the audio file
    audio = AudioSegment.from_file(file_path, format="mulaw")
    
    # Convert to 16-bit PCM
    audio = audio.set_sample_width(2)
    
    # Set the sample width to 16-bit
    audio = audio.set_sample_width(2)
    
    # Convert to PCM
    audio = audio.set_sample_width(2)
    
    # Get the output file path
    output_path = os.path.splitext(file_path)[0] + '.wav'
    
    # Export the audio to WAV format
    audio.export(output_path, format="wav")
    
    return output_path

def convert_mulaw_files(folder_path):
    for filename in tqdm(os.listdir(folder_path)):
        if filename.endswith('.wav'):
            if filename not in most_wanted:
                continue
            file_path = os.path.join(folder_path, filename)
            # Check if the audio file is 8-bit mu-law encoded
            audio_info = mediainfo(file_path)
            print(audio_info)
            codec_name = audio_info.get('codec_name')
            if codec_name == 'pcm_f32le':
                # Convert the file to 16-bit PCM WAV
                converted_file_path = convert_mulaw_to_pcm(file_path)
                print(f"Converted {filename} to 16-bit PCM WAV: {converted_file_path}")

folder_path = '/media/ahanaf/media-1/Datasets/ben10/ben10/16_kHz_train_audio'
convert_mulaw_files(folder_path)
```
