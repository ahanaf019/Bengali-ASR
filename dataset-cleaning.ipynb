```python
import pandas as pd

db_path = '/media/ahanaf/media-1/Datasets/ben10/ben10/'
```

```python
df = pd.read_csv(f'{db_path}/train.csv')
df
```

```python
df['transcripts_clean'] = df['transcripts'].apply(lambda x: x.replace('<>', ''))
df['transcripts_clean'] = df['transcripts_clean'].apply(lambda x: x.replace('\n', ''))
df['transcripts_clean'] = df['transcripts_clean'].apply(lambda x: x.strip())
df['len'] = df['transcripts_clean'].apply(lambda x: len(x))
df['transcripts_clean']
```

```python
df = df[df['len'] <= 400]
df = df[df['len'] >= 15]
```

```python
df
```

```python
df[df['len'] == 225]
```

```python
df = df.sample(df.shape[0])
```

```python
train_split = 0.9
n_samples = df.shape[0]

train_samples = int(train_split * n_samples)
```

```python
train_df = df[:train_samples]
```

```python
val_df = df[train_samples:]
```

```python
train_df.to_csv(f'{db_path}/cleaned_train.csv', index=False)
val_df.to_csv(f'{db_path}/cleaned_val.csv', index=False)
```
